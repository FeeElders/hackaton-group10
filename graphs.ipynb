```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data\KGM00357_monthly_flow.csv')
```

```python
# Ensure x-axis labels are rotated 90 degrees for better readability
for year, group in df.groupby('year'):
    group = group.sort_values('month')
    group['cumulative_flow'] = group['monthly_total_m3'].cumsum()
    plt.figure()
    sns.lineplot(data=group, x='month', y='cumulative_flow')
    plt.title(f'Cumulative Monthly Flow for year {year}')
    plt.xlabel('Month')
    plt.ylabel('Cumulative Flow (m3)')
    plt.xticks(rotation=90)
    plt.show()
```
